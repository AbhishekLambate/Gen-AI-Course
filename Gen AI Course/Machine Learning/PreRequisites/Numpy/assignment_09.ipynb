### Task 3: Important NumPy Mathematical Formulas
### Given:
### values = np.array([2, 4, 6, 8, 10])
### Calculate:
### 1. Square root of each element
### 2. Exponential of each element
### 3. Natural logarithm of each element
### 4. Sum of all elements
### 5. Cumulative sum of elements
### Use NumPy functions only.

```python
values = np.array([2, 4, 6, 8, 10])
```

```python
square_root=np.sqrt(values)
exp= np.exp(values)
log=np.log(values)
sums=np.sum(values)
cumsum=np.cumsum(values)
```

```python
print("Square Root:", square_root)
print("Exponential:", exp)
print("Natural Log:", log)
print("Sum:", sums)
print("Cumulative Sum:", cumsum)
```

### Task 5: Statistical Operations (Core Focus)
### Given:
### marks = np.array([78, 85, 90, 66, 72, 88, 95, 60])
### Calculate and print:
### 1. Mean
### 2. Median
### 3. Variance
### 4. Standard Deviation
### 5. Minimum & Maximum
### 6. Range (max - min)

```python
marks = np.array([78, 85, 90, 66, 72, 88, 95, 60])
```

```python
mean_v=np.mean(marks)
median_v=np.median(marks)
variance_v=np.var(marks)
standard=np.std(marks)
min_v=np.min(marks)
max_v=np.max(marks)
range_v=max_v - min_v

print(f"1. Mean: {mean_v}")
print(f"2. Median: {median_v}")
print(f"3. Variance: {variance_v}")
print(f"4. Standard Deviation: {standard}")
print(f"5. Minimum: {min_v}, Maximum: {max_v}")
print(f"6. Range: {range_v}")
```

### Task 6: Percentiles & Sorting
### Using the same marks array:
### 1. Sort the array
### 2. Find:
### 25th percentile
### 50th percentile
### 75th percentile
### 3. Count how many students scored above the average marks

```python
sort=np.sort(marks)
one=np.percentile(marks, 25)
two=np.percentile(marks, 50)
three=np.percentile(marks, 75)

avg=np.mean(marks)
count=np.sum(marks> avg)

print("Sorted Marks:", sort)
print("25th Percentile:", one)
print("50th Percentile (Median):", two)
print("75th Percentile:", three)

print("Average Marks:", avg)
print("Students scoring above average:", count)
```

### Task 7: Mini Use Case: Sales Analysis
### Given daily sales data: sales = np.array([1200, 1500, 900, 2000, 1800, 1700, 1600])
### Perform:
### 1. Total weekly sales
### 2. Average daily sales
### 3. Highest and lowest sales day
### 4. Standard deviation of sales
### 5. Identify days where sales were above average

```python
sales = np.array([1200, 1500, 900, 2000, 1800, 1700, 1600])
```

```python
total=np.sum(sales)
average=np.mean(sales)
high=np.max(sales)
low=np.min(sales)
std_dev=np.std(sales)
days=np.where(sales > average)[0]
print(f"Total: {total}")
print(f"Average: {average:.2f}")
print(f"Highest: {high}, Lowest: {low}")
print(f"Std Dev: {std_dev:.2f}")
print(f"Days above average (indices): {days}")
```
