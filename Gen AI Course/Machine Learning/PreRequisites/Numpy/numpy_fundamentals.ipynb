## Numpy

### Numpy is fundamental package for scientific computing in python. 
### It is a python library that provides multi dimentional array object and various derived objects such as masked arrays or matrices and an assortment of routines for fast operations on arrays including Mathematical, logical, shape manipulation, sortng, selecting, I/O, Discrete Fourier Transform, Basic linear algebra, Statical Operations, Random Simulation, etc.
### At the core of the numpy package is the n-dimentional array objects.


```python
## Why Numpy?
### 2008-09, Data science ->matlab, R language, 
### Data types in python -> list, Dictionary -> are slow
### Syntax -> python -> easy
### We need speed and leass memory to work with the huge amount of data.

#### Numpy -> c language -> wrapped -> python language
### -> Numpy -> syntax -> Easy for beginners -> Performing scientific calculations -> creating Multi Dimentional arrays -> speed is faster for applying operations -> less Memory 

### Libraries like scikit learn, Tensorflow and Pytorch require input data in numpy arrays or Tensors.
```

## Numpy arrays vs Python Sequences
### Numpy Arrays have fix size at creation, unlike python lists which grows dynamically. changing the size of numpy array will create a new array and delete the original .
### THe elements in a Numpy Array are required to be of the same data  type and that will be of the same size in memory.
### Numpy Arrays facilitate advanced mathematical and other types of operations on large number of data. Typically, such operations are executed more efficiently with less code that possible using python's built in sequences.

```python
## Stacking -> Joining the arrays together
## Horizontal Staking ->  Row by row joining is done -> hstack()
## Vertical Staking -> column by column joining done ->vstack()
```

```python
array1= np.arange(1,13).reshape(3,4)
```

```python
array1
```

```python
array2=np.arange(13,25).reshape(3,4)
```

```python
array2
```

```python
horizontal_array=np.hstack((array1,array2))
```

```python
horizontal_array
```

```python
vartical_array=np.vstack((array1,array2))
```

```python
vartical_array
```

```python
## Harizontal Split -->  np.hsplit(array,number of parts you want your array to be splitted at)
## Vartical Split --> np.vsplit(array,number of parts)
```

```python
np.hsplit(vartical_array,2)
```

```python
np.vsplit(vartical_array,2)
```

```python
np.vsplit(vartical_array,3)
```

```python
np.vsplit(vartical_array,4)
```
