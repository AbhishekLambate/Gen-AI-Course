## Functions

### 1. head() -> get the idea about the data that your dataset is storing as the top 5 rows, that means that starting from top we can get top 5 rows and along with thta if i want that nstead of top 5 rows i want to see a fixed no. of rows from the rows, that number i can pass inside the head() as a parameter

### 2. tail() -> provides the insights about the dataframe from the bottom so tail() can show us bottom 5 rows and again based on the value you will be passing as a parameter those many values you can get using the tail()

### 3. sample() -> provides the insights about the dataframe as the random values present inside the dataset and again based on how many values i want i can put the parameter as per that.

### 4. info() -> provides the idea  ->
### 1. about the data type of all columns
### 2. about the number of missing values of all columns
### 3. about dataframe storage in memory

### 5. describe() ->provides the idea as the mathematical summary of data -> min, max, count, mean, std, 25%, 50%, 75%
