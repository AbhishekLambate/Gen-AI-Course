## Pandas

### Pandas is the powerful library used for data manipulation and analysis. It helps us to work with data in Tabular format, just like how we deal with data and spreadsheet in excel or database.

### Pandas provides functionalities and attributes to clean, organize and analyze large amounts of data

### 1. Easy to use -> Works like excel inside python
### 2. Fast -> handles large data faster than manual methods
### 3.Clean Data -> helps clean, filter and transform messy data
### 4. Integrates well ->Matplotlib, Seaboarn
### 5. Ideal for ML and AI -> It is ideal for data science
