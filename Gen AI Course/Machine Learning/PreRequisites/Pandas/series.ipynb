## Pandas series is a column of data in a table like in excel or google sheets with labels called as indexes attached to each value.

```python
pd.Series(data=[40, 50, 60], index=['Bread', 'Milk', 'Butter'], name='Price of Grocery Items')
```

```python
## Series is a 1D Array that can hold any type of data like integer, float, strings (object), boolean, mixed data type.
```

## Data type -> Integers -> Data in a Series (Column) -> Quantity of item in the stock
## Data type -> float -> Data in a series (coulumn)  -> Price or Product Ratings
## Data type -> strings -> Column -> Product name
## Data type -> Boolean -> Column -> Is item in stock?

## Creating a series in multiple ways
### 1. List
### 2. Numpy Array
### 3. Dictionary
### 4. Custom Index
### 5. Scalar Value

```python
## Series from List
country=['India', 'China', 'USA', 'Russia']

pd.Series(country)
```

```python
batsman_runs=[42,65,73,33,86,135,234]
pd.Series(batsman_runs, dtype='int32')
```

```python
subjects=['maths','physics','chemistry','biology','english']
marks=[92,88,89,79,78]

pd.Series(data=marks,index=subjects,name='Subject Marks Data',dtype='int32')
```

```python
## Creating an series using a numpy array
values=np.arange(10,21)
pd.Series(values) 
```

```python
## Dictionary to series
## Dictionary is another data structure which stores the data in a key value pair and 
## just because it stores data in key value pair we dont have to explicitly mention the index
```

```python
data_items={'Maths': 98, 'English':78, 'Physics':82, 'Chemistry':89}
pd.Series(data_items)
```

```python
## Custom index to series

values=[1,2,3,4,5]
pd.Series(data=values, index=['a','b','c','d','e'])
```

```python
## Scalar values inside the Series

pd.Series(data=4, index=['a','b','c','d','e'])
```

### Mathematical functions in pandas

### 1. sum() -> sum of all values present in the series
### 2. product() -> product of all values in series
### 3. mean() -> Mean provides us the average value of the data column.
### 4. median() -> median provides the idea about the average of the data. for median , we need to sort the values in ascending manner and then the middle value after sorting the data

### 5. mode() -> mode provides the  idea that which value is getting repeated maximum
### 6. std() -> if i need to get idea that how the values inside the series are distributed from the averag or mean value then we can comes up with the standard deviation
### 7. variance() -> variance also provide the insights about the distribution of the data from the mean value and variance is the square of the standard deviation.
### standard_deviation=|mean - v1| + |mean- v2|..../total count of values
### variance=std**2 

### min() -> minimum value in the series
### max() -> maximum value in the series
