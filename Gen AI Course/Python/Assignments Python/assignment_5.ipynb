### Task 1: Create a Simple Module (math_utils.py)
### Create a module math_utils.py with the following functions:
### 1. add(a, b) - returns a + b
### 2. subtract(a, b) → returns a - b
### 3. square(n) returns n²
### In main.py, import this module in two different ways:
### import math_utils
### from math_utils import square
### Test all functions.

```python
import math_utils

print(math_utils.add(2,3))

print(math_utils.substract(4,2))

print(math_utils.square(5))
```
