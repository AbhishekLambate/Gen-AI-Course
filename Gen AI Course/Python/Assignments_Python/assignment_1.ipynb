```python
price_dict = {'laptop':70000,'pen':100,'mouse':550,'fan':2400,'keyboard':1200,'AC':45000}
print("price_dict is ",price_dict)
print(type(price_dict))
```

```python
price_dict['bag']=3000
print(price_dict)
```

```python
name=input("Enter the product name to remove : ")

if name in price_dict.keys():
    price_dict.pop(name)
    
    print(price_dict)
else:
    print("The entered product name not present in dict.")
    
    
```

```python
values=price_dict.values()
print(values)
```

```python
average=sum(values)/len(values)
print(average )
```

```python
min_value=min(values)
max_value=max(values)
print(min_value , max_value)
    
```

```python
min_product=None
max_product=None
min_price=float('inf')
max_price=float('-inf')
for key, value in price_dict.items():
    if value < min_price:
        min_price=value
        min_product=key
        
    if value > max_price:
        max_price=value
        max_product=key
        
print(f"The product {min_product} has minimum value i.e. {min_price} ")
print(f"The product {max_product} has maximum value i.e. {max_price} ")
```

```python
min_product = min(price_dict, key=price_dict.get)
max_product = max(price_dict, key=price_dict.get)

print(f"The product {min_product} has minimum value i.e. {price_dict[min_product]} ")
print(f"The product {max_product} has maximum value i.e. {price_dict[max_product]} ")
```

### Task 4: Combined Operations
### 1. Using the products list and price_dict, create a list of tuples named catalog where each tuple is (product_name, price, category).
### 2. From catalog, create a new dictionary category_to_products that maps each category to a list of product names in that category.
### 3. Print all products that belong to the category that has the maximum number of Products.

```python
catalog = [('laptop', 70000, 'electronics'), ('keyboard', 2000, 'electronics'), ('bat', 2500, 'sports'), ('panner', 500, 'food')]
print(type(catalog))
for c in catalog:
    print(c)
```

```python
category_to_products={}
for product_name, price, category in catalog:
    if category not in category_to_products:
        category_to_products[category] =[]
    category_to_products[category].append(product_name)
    
print(category_to_products)
```

```python
max_category=max(category_to_products, key=lambda category: len(category_to_products[category]))

print("category :", max_category)

for products in category_to_products[max_category]:
    print(products)
    # print(maximum)
```
