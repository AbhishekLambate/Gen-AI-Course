### Task 1: Discount Rules (if / elif / else)
### 1. Write a program that reads an integer order_amount from the user using input().
### 2. Apply the following discount rules and print the final amount:
### order_amount >= 2000 → 15% discount
### 1500 <= order_amount < 2000 10% discount
### 1000 <= order_amount < 1500 7% discount
### otherwise 0% discount
### 3. Ensure you convert input to a number and handle the case when the input is non-numeric by displaying an error message and exiting the program.
### Extra (optional): Add tax (fixed 5%) after discount and print subtotal, tax, and final total (still within conditionals and basic arithmetic).

```python
try:
    order_amount =int(input("Enter the amount : "))
except ValueError as e:
    print(e)
    print("Please enter numeric integer value")
else:
    if order_amount > 0:
        if order_amount >= 2000:
            subtotal= order_amount-(order_amount*0.15)
            tax=subtotal*0.05
            final_amount=subtotal+tax
            print("Subtotal", subtotal, "Tax", tax, "Final Amount ",final_amount)
        elif 1500 <= order_amount <2000:
            subtotal = order_amount - (order_amount* 0.10)
            tax=subtotal*0.05
            final_amount=subtotal+tax
            print("Subtotal", subtotal, "Tax", tax, "Final Amount ",final_amount)
        elif 1000 <=order_amount <1500:
            subtotal = order_amount - (order_amount* 0.07)
            tax=subtotal*0.05
            final_amount=subtotal+tax
            print("Subtotal", subtotal, "Tax", tax, "Final Amount ",final_amount)
        else:
            subtotal=order_amount
            tax=subtotal*0.05
            final_amount=subtotal+tax
            print("Subtotal", subtotal, "Tax", tax, "Final Amount ",final_amount)
    else:
        print("Please enter value greater than 0")
```

### Task 2: Process Multiple Orders (for loop)
### 1. Given a list of order amounts: rs = [1200, 2500, 800, 1750, 3000] use a for loop to apply the discount rules from Task 1 to each order and print a summary table
### showing: order_amount -> discount% -> final_amount.
### 2. Also compute and print the total revenue after discounts.
### Extra (optional): Print the number of orders that received a discount (discount > 0).

```python
rs=[1200, 2500,800,1750, 3000]
total_revenue=0
orders=0
for i in rs:
    if i > 0:
        if i >= 2000:
            final_amount= i-(i*0.15)
            discount=15
            orders+=1
            total_revenue+=final_amount
            print(f"{i} -> {discount}% -> {final_amount}")
        elif 1500 <= i <2000:
            final_amount = i - (i* 0.10)
            discount=10
            orders+=1
            total_revenue+=final_amount
            print(f"{i} -> {discount}% -> {final_amount}")
        elif 1000 <=i <1500:
            final_amount = i - (i* 0.07)
            discount=7
            orders+=1
            total_revenue+=final_amount
            print(f"{i} -> {discount}% -> {final_amount}")
        else:
            final_amount=i
            discount=0
            total_revenue+=final_amount
            
            print(f"{i} -> {discount}% -> {final_amount}")
print("Total revenue is : ", total_revenue)
print("Total number of orders that have  discount is ", orders)
```

### Task 3: User Menu (while loop + break/continue)

### 2- Show all orders and totals after applying discounts
### q - Quit
### 2. Use continue to re-show the menu after invalid input and break to exit on q.

```python
rs=[1200, 2500,800,1750, 3000]


while(True):
    print("1. Add order amount ")
    print("2. Show the orders ands totals")
    print("3. quit")
    ch = int(input("Entr your choice : "))
    
    if ch== 1:
        amount=int(input("Enter the order amount to add :"))
        
        rs.append(amount)
        print("Order added sucessfully")
    elif ch==2:
        total_revenue=0
        orders=0
        for i in rs:
            if i > 0:
                if i >= 2000:
                    final_amount= i-(i*0.15)
                    discount=15
                    orders+=1
                    total_revenue+=final_amount
                    print(f"{i} -> {discount}% -> {final_amount}")
                elif 1500 <= i <2000:
                    final_amount = i - (i* 0.10)
                    discount=10
                    orders+=1
                    total_revenue+=final_amount
                    print(f"{i} -> {discount}% -> {final_amount}")
                elif 1000 <=i <1500:
                    final_amount = i - (i* 0.07)
                    discount=7
                    orders+=1
                    total_revenue+=final_amount
                    print(f"{i} -> {discount}% -> {final_amount}")
                else:
                    final_amount=i
                    discount=0
                    total_revenue+=final_amount
                    
                    print(f"{i} -> {discount}% -> {final_amount}")
        print("Total revenue is : ", total_revenue)
        print("Total number of orders that have  discount is ", orders)
    elif ch==3:
        print("Thank You!!!")
        break
    else:
        print("Enter proper choice")
```

### Task 4: Loop Control with Conditions (break & continue)
### 1. Given a list of daily sales: daily = [200, 150, 0, 400, 50, -1, 300], iterate through it with a for loop and:
### If a day's value is -1, treat it as corrupted data and break the loop (stop processing).
### If a day's value is 0, treat it as a day with no sales and continue (skip adding to revenue).
### For valid positive sales, add to total_sales and print the running total.
### 2. Print the final total after the loop completes (or stops due to corrupted data).

```python
daily = [200, 150, 0, 400, 50, -1, 300]
total_sale = 0
for i in daily:
    if i== -1:
        print("There is corrupted data")
        break
    elif i== 0:
        print("There is no sales for the day")
        continue
    else:
        total_sale +=i
        print("The current total is", total_sale)
print("The final sale is", total_sale)
```
