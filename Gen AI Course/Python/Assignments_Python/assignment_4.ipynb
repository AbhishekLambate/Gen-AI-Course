### Task 4: Generate Summary Report from File
### Using only file reading operations:
### 1. Read all sales values from sales_data.txt.
### 2. Convert them into integers.
### 3. Calculate and print:
### Total Sales
### Highest Sale
### Lowest Sale
### Average Sale
### Do not use any advanced libraries.

```python
lists = []
total = 0
high = float('-inf')
low = float('inf')

with open('sales_data.txt', 'r') as f:
    content=f.read()
    print(content)
    numbers = content.splitlines()
    for num in numbers:
        lists.append(int(num))
    for i in lists:
        total+=i
        if high < i:
            high = i
        elif low > i:
            low =i
    ave = total/len(lists)        
    print("Total Sales is ", total)
    print("Highest Sale is ", high)
    print("Lowest Sales is ",low)
    print("Average sales is ", ave)
    
        
```

### Task 5: Create Product Info File (User Input)
### 1. Ask the user for 3 product names & their prices.
### 2. Write them into a new file products.txt in this format:
### 3. ProductName | Price
### 4. Read the file and print each line with proper formatting.

```python
with open('products.txt', 'w+') as f:
    for i in range (3):
        name=input("Enter the product name ")
        price=int(input("Enter the price for that product"))
        
        f.write(name + '|' + str(price) + '\n')
    f.seek(0)
    content=f.read()
    print(content)
```

### Task 7: Mini Project - Export Discounted Prices
### Create a dictionary:
### prices = {
### "Mouse": 500,
### "Keyboard": 800,
### "Monitor": 7000,
### "Pendrive": 400,
### "Camera": 5000
### }
### Ask the user for a discount percentage.
### Write discounted prices into discount_report.txt using:
### Product | Original Price | Discounted Price
### After writing, read the file and print it to the terminal.
### Extra (optional): Write a summary at the bottom of the file:
### Total Items: X
### Average Discounted Price: Y

```python
prices = {
 "Mouse": 500,
 "Keyboard": 800,
 "Monitor": 7000,
 "Pendrive": 400,
 "Camera": 5000
 }
items=len(prices)
dis_price=0

with open('discount_report.txt' , 'w+') as f:
    for key, value in prices.items():
        dis=int((input(f" Enter discount for {key} " )))
        disc= value - (value * dis/100)
        dis_price += disc
        f.write(key + '|' + str(value) + '|' + str(dis) + '|' + str(disc) + '\n')
        
    f.seek(0)
    content=f.read()
    print(content)
    
    print("Total items are ",items)
    print("Average Discounted Price ", (dis_price/items))
        
        
```
