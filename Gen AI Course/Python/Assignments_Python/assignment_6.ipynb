### Task 2: Bill Calculator with Error Handling
### Given a list of product prices:
### prices = [120, 350, 'abc', 500, -200, 800]
### Write code that:
### 1. Iterates through the list.
### 2. Tries to add only valid (positive numerical) prices to the total.
### 3. Handles:
### TypeError if value is not a number
### Custom exception using raise ValueError("Negative price not allowed")
### 4. Prints the running total.
### Expected behavior:
### Skip invalid items but continue processing.

```python
prices = [120, 350, 'abc', 500, -200, 800]
total = 0
try:
    for i in prices:
        try:
            if i < 0:
                print("Negative price not allowed")
            else:
                total= total + i
                print(total)
        
        except TypeError as e1:
            print(e1)
        continue
    
except Exception as e:
    print(e)
            
```

### Task 3: Custom Exception: Age Validator
### 1. Write a function check_age(age) that:
### Raises a custom exception ValueError("Age must be between 1 and 120") if age is out of range.
### 2. In your main code:
### Take age input from the user.
### Use try-except to catch and print the custom error message.

```python
def check_age(age):
    if 1 > age or age > 120:
        raise ValueError("Age is out of range")
        
try:
    age=int(input("Enter the age "))
    check_age(age)
    print("valid age")
except ValueError as v:
    print(v)
```

### Task 4: File Reader with Exception Handling
### 1. Ask the user for a filename.
### 2. Try to open and read the file.
### 3. Handle:
### FileNotFoundError
### PermissionError
### 4. If successful, print first 3 lines of the file.
### 5. Use finally to print:
### "File operation attempted."

```python
name=input("Enetr the file name ")
try:
    with open(name, 'r') as f:
        for i in range(3):
            content=f.readline()
            if not content:
                break
            print(content.strip())
except FileNotFoundError as e1:
    print("FileNotFoundError ", e1)
except PermissionError as e2:
    print(e2)
    
finally:
    print("File operation attempted")
```

### Task 5: Mini Program: Safe Shopping Cart
### Create a program that:
### 1. Has a cart list cart =
### 2. Runs a loop asking user to enter prices.
### 3. Stops when user enters 'q'.
### 4. Inside the loop:
### Convert input to a float
### Handle ValueError if user enters invalid input
### Raise custom exception if price is negative
### 5. At the end, print:
### Total items
### Total bill

```python
cart =[]
total_bill=0
while(True):
    print(cart)
    print("1. Want to enter price")
    print("2. Quit")
    
    ch=int(input("Enetr the choice "))
    
    if ch==1:
        try:
            price=int(input("Enter the price "))
            if price < 0:
                raise ValueError ("You entered negative value")
            else:
                cart.append(price)
        except ValueError as v:
            print(v)
    elif ch==2:
        print("Ending it now printing result")
        items =len(cart)
        for i in cart:
            total_bill+=i 
        print("Total items ", items)
        print("Total bill ",total_bill)
        break
    else:
        print("Enter correct choice")
    
```
