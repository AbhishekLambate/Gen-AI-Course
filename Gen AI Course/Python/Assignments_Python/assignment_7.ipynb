#### Task 1: Basic Class & Object Creation
#### Create a class Product with:
#### Attributes: name, price, category
#### Method: get_info() prints product details
#### Create two objects and call get_info().
#### Extra (optional): Add a method apply_discount(percent) that returns the discounted price.

```python
class Product:
    def __init__(self, name, price, category):
        self.name=name
        self.price=price
        self.category=category
        
    def get_info(self):
        print(f"The name of product is {self.name}. The price of product is {self.price}. The category of product is {self.category}")
        
    def apply_discount(self, percent):
        new_price= self.price - (self.price* percent/100)
        print(f"The new price after discount of {percent} % is {new_price}")
```

```python
pen=Product('pen', 20, 'Education')
pen.get_info()
```

```python
laptop=Product('Laptop', 450000, 'Electronics')
laptop.get_info()
```

```python
pen.apply_discount(5)
```

#### Task 2: Constructor & Encapsulation
#### Modify the Product class:
#### Make price a private attribute (price).
#### Create getter & setter methods:
#### get_price()
#### set_price(new_price) → should update only if new_price > 0
#### Test modifying price using the setter.

```python
class Product:
    def __init__(self, name, price, category):
        self.name=name
        self.__price=price
        self.category=category
        
        
        
    ## Getter
    def get_price(self):
        # print(f"The price of product is {self.__price}")
        return self.__price
        
    ## Setter
    def set_price(self,new_price):
        if new_price > 0:
            self.__price=new_price
            
        print(f"The updated price is {self.__price}")
        
        
    def get_info(self):
        print(f"The name of product is {self.name}. The price of product is {self.get_price()}. The category of product is {self.category}")
        
    def apply_discount(self, percent):
        new_price= self.get_price() - ((self.get_price()) * (percent/100))
        print(f"The new price after discount of {percent} % is {new_price}")
```

#### Task 3: Inheritance (Single-Level)
#### Create a subclass Electronic Product that inherits from Product.

#### Override the get_info() method to include warranty info.
#### Create an object and demonstrate inheritance + overriding.

```python
class Electronic(Product):
    def __init__(self, name, price, category, warrenty_years):
        super().__init__(name, price, category)
        self.warrenty_years=warrenty_years
        
    def get_info(self):
        return f"The name of product is {self.name}. The price of product is {self.get_price()}. The category of product is {self.category} and the warrenty of {self.warrenty_years} years."
```

```python
class Laptop(Product):
    def __init__(self, name, price, category):
        super().__init__(name, price, category)
        
    def get_info(self):
        return f"The name of laptop is {self.name}.  \n The price of it is {self.get_price()}. \n The category of it is {self.category}"
    
class Mobile(Product):
    def __init__(self, name, price, category):
        super().__init__(name, price, category)
        
    def get_info(self):
        return f"The name of Mobile is {self.name}.  The price of it is {self.get_price()}. The category of it is {self.category}"
     
```

#### Task 5: Abstraction (Using Abstract Base Class)
#### Create an abstract class Payment with abstract method:
#### process_payment(amount)
#### Then create two subclasses:
#### CreditCardPayment
#### UPIPayment
#### Both override process_payment() with simple print statements.
#### Test all classes.

```python
from abc import ABC, abstractmethod

class Payment(ABC):
    
    @abstractmethod
    def process_payment(self, amount):
        pass
    
class CreditCardPayment(Payment):
    def process_payment(self, amount):
        print(f"the amount of creadit card is {amount}")
        
class UPIPayment(Payment):
    def process_payment(self, amount):
        print(f"the amount of UPI is {amount}")
```

```python
cre=CreditCardPayment()
```

```python
cre.process_payment(20)
```

```python
upi=UPIPayment()
```

```python
upi.process_payment(100)
```

#### Task 7: Mini Project: Simple Inventory System (OOP Only)
#### Create two classes:
#### Class: Inventory
#### Attributes:
#### products list to store product objects
#### Methods:
#### add_product(product)
#### remove_product(name)
#### get_total_value() sums prices of all products
#### show_all_products() prints info for each product
#### Class: Store
#### Attributes:
#### store_name
#### inventory an Inventory object
#### Methods:
#### add_new_product() - takes input & creates Product object
#### show_summary() prints total items & value
#### no file handling, no exceptions, no packages. Important: Use only OOP concepts
#### Test the system by:
#### 1. Creating a Store object
#### 2. Adding 3 products
#### 3. Showing summary
#### 4. Using_add_ to combine prices of two products

```python

class Product:
    def __init__(self, name, price):
        self.name=name
        self.__price=price
        
    def get_price(self):
        return self.__price
    
    def __str__(self):
        return f"product({self.name}, {self.__price})"
    
    def __add__(self, other):
        return self.get_price() + other.get_price()
    
    
class Inventory:
    def __init__(self):
        self.products=[]
        
    def add_product(self, product):
        self.products.append(product)
        
    def remove_product(self, name):
        self.products=[
            product for product in self.products
            if product.name !=name
        ]
        
    def get_total_value(self):
        total = 0
        
        for product in self.products:
            total+= product.get_price()
        return total
    
    def show_all_products(self):
        for product in self.products:
            print(product)
            
class Store:
    def __init__(self, store_name):
        self.store_name=store_name
        self.inventory=Inventory()
        
    def add_new_product(self, name, price):
        product = Product(name, price)
        self.inventory.add_product(product)
        
    def show_summary(self):
        print(f"Store Name: {self.store_name}")
        print(f"Total Products: {len(self.inventory.products)}")
        print(f"Total Inventory Value: {self.inventory.get_total_value()}")

            
    
```

```python
store = Store("Tech Store")

store.add_new_product("Dell Laptop", 50000)
store.add_new_product("Moto G96", 20000)
store.add_new_product("Mouse", 1000)

print("All Products:")
store.inventory.show_all_products()

print("\nStore Summary:")
store.show_summary()
```
