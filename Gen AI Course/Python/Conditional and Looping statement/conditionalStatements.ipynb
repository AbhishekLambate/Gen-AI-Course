## Conditional Statements

### Condtional statements works like real life scenario  where if a certain condition is meeting then a statement of code or block of code gets executed

```python
print("hi")
```

```python
# eligibility for driving on meeting certain condition

# conditional statements -> if statement where if a condition is meeting then a block of code get executed

age=int(input("Enter your age: "))

if age>=18:
    print("Yes, You are eligible for driving")
else:
    print("Sorry, you are not eligible for driving")
```

### Conditional statements:
#### if statement
#### elif statement
#### else statement

```python
# figure out the largest number among three numbers
a=int(input("Enter the first number: "))
b=int(input("Enter the second number: "))
c=int(input("Enter the third number: "))

if a>=b and a>=c:
    print("The largest number is ",a ,"among other numbers", b,"and",c)
elif b>=a and b>=c:
    print("The largest number is ",b ,"among other numbers", a,"and",c)
else:
    print("The largest number is ",c ,"among other numbers", a,"and",b)
```

## Nested Conditpional statements

### Nested if statement is an if statement inside another if statement. It is used to check multiple conditions in a hierarchical manner.

```python
# Compare the heights of three friends and figure out the tallest one

friend1=int(input("Enter the height of friend 1: "))
friend2=int(input("Enter the height of friend 2: "))
friend3=int(input("Enter the height of friend 3: "))
if friend1>friend2:
    if friend1>friend3 or friend1>friend3:
        print("Friend 1 is the tallest among three friends")
elif friend2>friend1:
    if friend2>friend3 or friend2>friend3:
        print("Friend 2 is the tallest among three friends")
else:
    print("Friend 3 is the tallest among three friends")
```

```python
# check whether the year is leap year or not
year=int(input("Enter the year: "))
if year%4==0:
    if year%100==0:
        if year%400==0:
            print("The year is a leap year")
        else:
            print("The year is not a leap year")
    else:
        print("The year is a leap year")
else:
    print("The year is not a leap year")
```
