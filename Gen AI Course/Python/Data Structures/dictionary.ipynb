### Nested Dictionary

### Nested Dictionary will be used for keeping a dictionary inside another dictionary

```python
# I want to store the information for multiple students like name, marks, grade

nesteddict={"Student1":{"Name":"Abhi","Marks":85,"Grade":"A"},
            "Student2":{"Name":"omii","Marks":88,"Grade":"A"},
            "student3":{"Name":"shubz","Marks":95,"Grade":"A+"}}
nesteddict
```

```python
# iterate over these nested dictionary

for outer_key, outer_value in nesteddict.items():
    print(f"Outer Key: {outer_key} and outer value: {outer_value}")
    for inner_key, inner_value in outer_value.items():
        print(f"  {inner_key}: {inner_value}")
```

### Dictionary comprehension

### Dictionary comprehension helps us with using looping statements along with conditional statements inside our dictionary. It is a concise way to create dictionaries.


### Syntax for Dictionary comprehension:

### {epression for item in iterable if condition}


```python
comprehensive={i**2 for i in range(11)}
```

```python
comprehensive
```

```python
comprehensive={i: i**2 for i in range(11)}
comprehensive
```

```python
even_squares={i: i**2 for i in range(1,11) if i%2==0}
even_squares
```
