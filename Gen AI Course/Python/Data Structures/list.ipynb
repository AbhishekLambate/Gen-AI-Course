## List Comprihension

### We are going to use looping startements inside the list so that we can store  the value inside the list in a iterable format and in list comprehension conditional statements are allowed as well

## Syntax for list comprehension

### [expression for item in iterable if condition]
### epression-> what we want to print
### for the item in iterable-> looping statement
### if condition-> coditional statement

```python
# I need to store values from 1 to 50 inside the list

list_items=[i for i in range(1,51)]
list_items
```

```python
squares=[i**2 for i in range(1, 12)]
squares
```

```python
# if i want square values which are event

even=[i**2 for i in range(1,12) if i%2==0]
even
```

```python
# Functions in list comprehension

items=["Abhishek", "Apple", "Banana", "car", "Laptop"]

length=[len(word) for word in items]
length
```
