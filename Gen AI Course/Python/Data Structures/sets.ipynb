## Sets

### Sets are going to be data structure which can store values inside it within the curly braces {} and known for storing the values which are non duplicates and unique

```python
a=set()
type(a)
```

```python
sets={1,2,3,4,5}
type(sets)
```

## methods in sets


```python
#add()-> adds an element to the set from last
sets.add(6)
print(sets)
```

```python
sets.add(7)
print(sets)
```

```python
#sets only consists of unique values inside it and non duplicate values 
sets.add(3)
print(sets)
```

```python
# user is asking to add values inside a set which must be unique and non duplicate

set_values={1,2,3,4,5}

for i in range(0,21):
    set_values.add(i)
print(set_values)
```

```python
fruits={"apple","banana","grapes","orange","tomato"}
fruit_vegetables={"apple","potato","onion","tomato"}
for item in fruit_vegetables:
    fruits.add(item)
print(fruits)
```
