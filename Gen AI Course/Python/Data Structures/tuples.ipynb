## Tuples

### Tuples are the type of data structure which can store the data in a way just like list but tuples are not mutable

### tuples will be created inside these brackets()

```python
a=tuple()
type(a)
```

```python
values=(1,2,3,4,5)
print(values)
```

```python
### Tuples are like lists but they are immutable. Once you create a tuple, you cannot change its contents. They are defined using parentheses `()` instead of square brackets `[]`.

tuple1 = (1, 2, 3)
print(tuple1[0])  # Accessing the first element of the tuple
```

```python
tuple1[0] = 10  # This will raise a TypeError because tuples are immutable
```

```python
### tuples are also capable of storing a data of any data type.
data = (1, "Hello", 3.14, [1, 2, 3], (4, 5))
print(data)
```

```python
### tuples are gong to be ordered or are going to be indexed. We can access the elements of a tuple using their index, just like we do with lists. The index starts at 0 for the first element, 1 for the second element, and so on.

data[:]
```

```python
data[1:]
```

```python
data[:-1]
```

```python
data[::-1]
```

```python
data[1:3]
```

```python
## Nested List
lists=[[1,2,3],[4,5,6],[7,8,9,13],[10,11,12,14,15]]

print(lists[0][:])  # Accessing the first element of the first list
```

```python
# Tuples inside a list
list_of_tuples = [(1, 2), (3, 4), (5, 6)]
print(list_of_tuples[0][:])  # Accessing the first tuple
```

```python
# Nested Tuple
tuples=((1,2,3),(4,5,6),(7,8,9,13),(10,11,12,14,15))
tuples[0][:]
```

```python
# iterating over a tuple inside tuple
for i in tuples:
    for j in i:
        print(j, sep=" ")
    print()
```
