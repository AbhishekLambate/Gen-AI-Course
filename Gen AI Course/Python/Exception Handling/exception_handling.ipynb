### Exception handling ia a way in which if we are getting the error in programming or if the user entered input cause error, then instead of throwing an error we can raise an exception which can let the user know that the given input is not correct or the input can come up as an error.

### As we dont want to stop the execution of the program  so we raise the error which will let the user know that current input is not correct but the program execution wont stop

### Exceptions are the events that disrupt the normal flow of a program. They occur when an error is encountered during program execution

### Types of Exceptions:

### ZeroDivisionError  -> Dividing by zero
### FileNotFoundError -> File not found
### ValueError   -> Invalid value
### TypeError   -> Invalid Type

### try -> try block is going to have the operations where we can expect an error to occur
### except -> which is used to handle the error in a way such that the program execution shouldn't stop
### else -> It will be consisting the code which we want to run if the try block operation is working without throwing any error
### finally -> block gets executed once we are done with Everything and even if we are getting an error or not getting an error still the code inside the finally block executed

### if try -> code throws an error -> Except -> code will be running
### if try -> code not throws an error -> Else -> code will be running
