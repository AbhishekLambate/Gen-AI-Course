```python
## open()-> opening the file
## first Parameter-> file name
## second parameter-> mode of operation
## Mode -> It decides the operation we want to perform on the file
'''r-> read -> Reading the content of the file
w-> write -> Writing the content into the file
 
 
r+ -> Read the content first and then writing the content into the file
w+ -> Write the content first and then reading the content from the file'''
```

```python
## Reading the content of the file
## read()-> It reads the content of the file and returns it as a string
```

```python
f=open("example.txt", "r")
content=f.read()
print(content)
f.close()
```

### Writing the content 
### write() ->If the file is not present it creates a new file and then writes the content into it. If the file is already present then it overwrites the content of the file.
### append() -> It adds the data from the end of the file

### mode w-> if mode is w then content will be overwrite
### mode a-> if mode is a then content will be added at end

```python
## write() -> It writes the content into the file. If the file is not present it creates a new file and then writes the content into it. If the file is already present then it overwrites the content of the file.

f=open("example.txt", "w")
f.write("Hello World! we are learning file handling in python")
f.close()
```

```python
## append() -> It is used to append the content into the file. If the file is not present it creates a new file and then writes the content into it. If the file is already present then it appends the content into the file.

f=open('example.txt', 'a')
f.write("\n we are appending content in file")
f.close()
```

### with -> it will be used for first of all opening the file as a object and there is no requirement for closing the file as well

```python
with open("example.txt", "r") as f:
    content=f.read()
    print(content)
```

```python
with open("example.txt","w") as f:
    f.write("we are exploring the with command of file handling")
```

```python
with open('example.txt', 'a') as f:
    f.write("\n we are appending content in file using with command")
```

```python
# Write the content into the file line by line using writelines() method

lines=["This is line 1\n", "This is line 2\n", "This is line 3\n"]
with open("example.txt", "w") as f:
    f.writelines(lines)
```

```python
# Mode -> write the content into the binary file -> "wb"
# Mode -> read the content from the binary file -> "rb"

data=b"Hello World! we are learning file handling in python"
with open("example.txt", "wb") as f:
    f.write(data)
```

```python
with open("example.txt","rb") as f:
    d=f.read()
    print(d)
```

```python
# Reading the total lines, total words and total characters from the file -> readlines()

def counts(file):
    with open(file,'r') as f:
        lines=f.readlines()
        total_lines=len(lines)
        total_words= sum(len(line.split()) for line in lines)
        total_char=sum(len(line) for line in lines)
    return total_lines, total_words, total_char

file_path='total_data.txt'
Total_Lines, Total_Words, Total_charactrs = counts(file_path)
print("Total Lines :", Total_Lines)
print("Total Wordes :", Total_Words)
print("Total Characters :", Total_charactrs)
```
