```python
# getcwd() -> It is capable of providing us with a idea about a current working directory
# getcwd -> get current working directory
```

```python
import os
```

```python
current_path=os.getcwd()
print("The current working directory is : ", current_path)
```

```python
# mkdir() -> make directory -> It will help us out with creating the directory
```

```python
new_path="New_Folder"
os.mkdir(new_path)
print(f'The new directory {new_path} is created sucessfully')
```

```python
## listdir() -> Listing all of the directories present -> List Directories

All_items=os.listdir(".")
print(All_items)
```

```python
## Joining 2 paths together

## os.path.join() -> It will help us to joining multiple paths together

dir_name="Folder_data"
file_name="File_data"

full_path=os.path.join(dir_name, file_name)

print(f"The full path is : {full_path}")
```

```python
# I want to join multiple path with the current path

current_path=os.path.join(os.getcwd(),dir_name,file_name)
print(current_path)
```

```python
## os.path.exists()

given_path='/home/ubuntu/Documents/Abhi_Important/Gen-AI-Course/Gen AI Course/Python/app.py'
if os.path.exists(given_path):
    print("yes, path exists")
else:
    print("NO, Path does not exists")
```

```python
# os.path.isfile() -> This checks that whether the current path is file or not
# os.path.isdir() -> This checks that whether the current path is dir or not

path="total_data.txt"

if os.path.isfile(path):
    print("The current path belongs to file")
elif os.path.isdir(path):
    print("The current path belongs to directory ")
else:
    print("The current path is neither file nor directory")

```

```python
path="New_Folder"

if os.path.isfile(path):
    print("The current path belongs to file")
elif os.path.isdir(path):
    print("The current path belongs to directory ")
else:
    print("The current path is neither file nor directory")

```

```python
# The complete path -> Absolute path -> os.path.abspath()

# os.path.abspath() -> It provides us with the complete path of the absolute path of the file or directory


path="total_data.txt"

full_path=os.path.abspath(path)
print(full_path)
```
