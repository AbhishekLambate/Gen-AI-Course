## Recursve Function

## 

### By recursive function we can say that a function is going to call again and again

```python
# factorial of a number
def factorial(num):
    if num==0 or num==1:
        print(f"The factorial of {num} is 1")
    else:
        fact=1
        for i in range(1,num+1):
            fact=fact*i
        print(f"The factorial of {num} is {fact}")
        
num=int(input("Enter number "))

factorial(num)
    
```

```python
# factorial of number by recursive

def factorial(num):
    if num==0 or num==1:
        return 1
    else:
        return num* factorial(num-1)
    
factorial(5)
```

### Function with multiple parameters

```python
## Function for addition of two numbers

def addition(num1,num2):
    return num1+num2

num1=int(input("Enter first number"))
num2=int(input("Enter second number"))
print(f"The addition of {num1} and {num2} is {addition(num1,num2)}")
```

### The function with variable length of arguments

#### Positional arguments-> positional arguments are being used so that we can pass multiple values inside our function as an argument irrespective of the number of values we have mentioned as the parameter inside our function


```python
# function to add multiple numbers irrespective of how many numbers we are passing

def addition(*num):
    sum=0
    for i in num:
        sum=sum+i
    print("The addition of numbers is ",sum)
    
addition(1,2,3,4)
```

### Keyword arguments in function-> takes the input as a keyword values and irrespective of any number of keyword values you are going to pass it wont matter  

```python
def data(**user_data):
    for key,value in user_data.items():
        print(f"{key} : {value}")
        
data(name="Abhishek",age=25,city="Delhi",country="India")
```

### Want to pass keyword value pairs as your input along with positional arguments as well

```python
def variable(*positional, **keyword):
    for i in positional:
        print(f"positional values are {i}")
        
    for key,value in keyword.items():
        print(f"keyword values are {key} : {value}")
        
variable(1,2,3, name="Abhishek",age=25,city="Delhi",country="India")
```
