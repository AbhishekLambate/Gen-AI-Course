```python
# To create a function that converts temperature from celcius to fahrenheit or fahrenheit to celcius
# 1 deg celcius = 9/5 +32 fahrenheit
# 1 fahrenheit = 5/9 -32 deg celcius

def temp_convert(temp, unit):
    if unit=="C" or unit=="c":
        temp=temp*9/5 +32
        return (str(temp) +"F")
    elif unit=="f" or unit=="F":
        temp=temp*5/9 -32
        return (str(temp) +"C")
    else:
        print("Please enter proper value and unit")
    
temperature=int(input("Enter the temperatur : "))
unit=input("Enter the unit : ")
converted=temp_convert(temperature , unit)
print(f"The converted temperatur is {converted}")


```

```python
# define a function that would behave as a password Strength checker. It should check for the following:
# 1. The password should be at least 8 characters long
# 2. The password should contain at least one uppercase letter, one lowercase letter, one digit and one special character
# 3. The password should not contain any spaces 

def password_strength(password):
    if len(password) < 8:
        return "Password should be at least 8 characters long"
    elif not any(char.isupper() for char in password):
        return "Password should contain at least one uppercase letter"
    elif not any(char.islower() for char in password):
        return "Password should contain at least one lowercase letter"
    elif not any(char.isdigit() for char in password):
        return "Password should contain at least one digit"
    elif not any(char in "!@#$%^&*()-_=+[]{}|;:'\",.<>?/" for char in password):
        return "Password should contain at least one special character"
    elif " " in password:
        return "Password should not contain any spaces"
    else:
        return "Password is strong" 
    
password=input("Enter the password for check : ")
resp=password_strength(password)
print(f"For entered password : {password} response is : {resp}. ")
```

```python
# check if the given string is a palindrome or not. 
# A palindrome is a word, phrase, number, or other sequence of characters that reads the same forward and backward (ignoring spaces, punctuation, and capitalization).

def is_palindrome(string):
    string=string.replace(" ", "").lower()
    return string == string[::-1]

palindrome_string=input("Enter the string to check for palindrome : ")
if is_palindrome(palindrome_string):
    print(f"{palindrome_string} is a palindrome.")
else:
    print(f"{palindrome_string} is not a palindrome.")
```

```python
# calculate the total cost of items in a shopping cart. 
# The function should take a list of items, where each item is a dictionary with keys "name", "price", and "quantity". 
# The function should return the total cost of all items in the cart.

def calculate_total_cost(cart):
    total_cost = 0
    for item in cart:
        total_cost += item["price"] * item["quantity"]
    return total_cost
shopping_cart = [
    {"name": "apple", "price": 5, "quantity": 4},
    {"name": "banana", "price": 3, "quantity": 6},
    {"name": "orange", "price": 7, "quantity": 3}
]
total_cost = calculate_total_cost(shopping_cart)
print(f"The total cost of items in the shopping cart is: {total_cost:.2f}")
```
