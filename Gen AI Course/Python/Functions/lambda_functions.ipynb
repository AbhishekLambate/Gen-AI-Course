## Lambda Functions

### Lambda functions are the one liner functions defined by the lambda keyword

### Syntax for lambda function:
#### lambda arguments: expression
#### lambda -> Keyword which simply let the python interpreter know that the current function is our lambda function
#### Arguments -> Input values for our function
#### Expression -> Logic (function body)

####  lambda x:x**2

#### They can have any number of arguments but only one epression or only one logic
#### They are used for short operations or as arguments to higher order functions

#### It can be treated as a function without name and dont even required a return statement

```python
print("hi")
```

```python
# function which calculate the square of a number
lambda_function=lambda x:x**2

type(lambda_function)

print(lambda_function(5))

```

```python
# lambda function to calculate the sum of two numbers
sum=lambda a,b:a+b
print(sum(3,5))
```

```python
product=lambda a,b:a*b
print(product(3,5))
```
