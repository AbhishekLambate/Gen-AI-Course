## Map Functions

### Map function applies a given function operation to all tems in an iterable or a list and return a map object(an iterator) 

### Map function is particularly useful for transforming tha data or applying operation on all items of the iterable

### map() is going to help us out with avoiding unnecessory lines of code and we can apply operation on all valuea or all items in an iterable

### Syntax for map()

### list(map(function, iterable))

```python
print("hi")
```

```python
# Squring all items of a list
def square(items):
    squ=[]
    for i in items:
        squares=i**2
        squ.append(squares)
    return squ

item=[1,2,3,4]
print(square(item))        
```

```python
lambda_function=lambda x:x**2
item=[1,2,3,4]
print(list(map(lambda_function,item)))
```

```python
### Map function with basic function in it

def square(items):
    return items**2

print(list(map(square,item)))
```

### Whenever you are going to pass a general function inside your map function as the parameter then you dont have to use parenthesis

### While mapping the values of one iterable  with another iterable we can actually apply the operation in between the values of the two iterables in a synchronous way
### [1,2,3,4,5] <-> [6,4,2,1,3] -> Addition->[7,6,5,5,8]   

```python
# Map functions with multiple iterables
num1=[1,2,3,4]
num2=[5,6,7,8]

lambda_function=lambda x,y:x+y

print(list(map(lambda_function,num1,num2)))
```

```python
## Map() to convert all strings values items in an iterable into an integer
num=["1","2","3","4"]

print(list(map(int,num)))
```

```python
# convert all string values to upper case

names=["Abhishek", "Amit", "Omkar"]

print(list(map(str.upper, names)))

```

```python
# Retrive the values inside the dictionary using map

employee=[{"name":"Abhishek","age":"24","city":"pune"},
          {"name":"Amit","age":"25","city":"mumbai"},
          {"name":"Omkar","age":"25","city":"baramati"}]

def names(items):
    return items["name"]

print(list(map(names,employee)))
```
