```python
## CSV library -> Most of the time we are going to work with csv files so that knowledge with csv library functionality matters

import csv
```

```python
with open("example.txt", "w", newline="") as f:
    writer=csv.writer(f)
    writer.writerow(['name', 'age'])
    writer.writerow(['Abhishek',24])
```

```python
with open('example.txt','r') as f:
    reader=csv.reader(f)
    for row in reader:
        print(row)
```

```python
## JSON library

## when we are going to work with Nosql Database data then the data inside the nosql database is going to be their in the json format and along with that when we are going to work with the datasets as well then again these dataset will be consisting of data in the json format itself
```

```python
## when we get json data , we get in a json string format so by using json library we can convert our string data into dictionary format for serializing our data

import json


data={"name": "Abhishek", "Age":25}
json_str = json.dumps(data)
print(json_str)
print(type(json_str))
```

```python
json_dict= json.loads(json_str)
print(json_dict)
print(type(json_dict))
```

```python
## Regular Expression Library

## It is used to retrive the info based on specific pattern
```

```python
import re

text="Hello my name is Abhishek and my password is Abhi123"

pattern = r'\d+'
match=re.search(pattern, text)
print(match.group())
```
