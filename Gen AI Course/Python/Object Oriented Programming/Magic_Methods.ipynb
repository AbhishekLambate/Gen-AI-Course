## Magic Methods

## Magic methods are also called as dunder methods because they have double underscore __ before and after their name and they are special methods in python that allows us to customize how object behave


## these methods can help us define:
## 1. Object Intialization (__init__)
## 2. String Rrepesentation (__str__, __repr__)
## 3. Arithimatic Operations (__add__, __sub__, etc)
## 4. Comparison Operations (__eq__, __lt__)
## 5. length, indexing, iteration (__len__, __getitem__, etc)

## Why to use Magic methods
## 1. Makes objects behave like built in types (eg, lists, integers)
## 2. Improves Readability by allowing basic operations object1 + object2
## 3.Simplifies code by replacing verbose function calls with natural syntax

## print(object) -> __str__ -> what is returned when print(object) is used
## Arithmatic operations can be done among the object by using Magic methods

```python
## __add__

class Avenger:
    def __init__(self, name, strength):
        self.name=name
        self.strength=strength
        
    def __add__(self, other):
        return Avenger(self.name + "&" + other.name, self.strength + other.strength)
    
    def __str__(self):
        return (f"{self.name} with combine strength {self.strength}")
    
ironman=Avenger("Iron Man", 80)
thor = Avenger("Thor", 90)

team= ironman + thor
print(team)
```
