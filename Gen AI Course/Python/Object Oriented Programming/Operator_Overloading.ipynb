## Operator Overloading

## Operator Overloading allows us to define how operators (+,_,/,*,etc) works with user defined objects.
## Why is Operator Overloading useful?
## Makes custom objects behave like built in types (numbers, strings, lists)
## Improves code readability by using natural syntax  (object1 + object2) instead of object.add(object2)


## Operator -> + -> method name -> __add__(self, other)
## Operator -> - -> method name -> __sub__(self, other)
## Operator -> * -> method name -> __mul__(self, other)
## Operator -> / -> method name -> __truediv__(self, other)
## Operator -> == -> method name -> __eq__(self, other)
## Operator -> > -> method name -> __gt__(self, other)


```python
class Avenger:
    def __init__(self, name, power):
        self.name=name
        self.power=power
        
    def __add__(self, other):
        return Avenger(self.name + " & " + other.name, self.power + other.power)
    
    def __sub__(self, other):
        return Avenger(self.name, self.power - other.power)
    
    def __mul__(self, other):
        return Avenger(self.name, self.power * other.power)
    
    def __truediv__(self, other):
        return Avenger(self.name, self.power/other.power)
    
    def __str__(self):
        return f"{self.name} with Power {self.power}"
    
    
```

```python
thor=Avenger("Thor", 100)
hulk=Avenger("Hulk", 120)
```

```python
team=thor + hulk

print(team)
```

```python
class AIModel:
    def __init__(self, name, accuracy):
        self.name=name
        self.accuracy=accuracy
        
    def __gt__(self, other):
        return self.accuracy > other.accuracy
    
    def __lt__(self, other):
        return self.accuracy <other.accuracy
    
    def __eq__(self, other):
        return self.accuracy == other.accuracy
    
```

```python
gpt4=AIModel("GPT-4", 98)
bard=AIModel("Google Bard", 96)
```

```python
print(gpt4 > bard)
```

```python
print(gpt4 < bard)
```

```python
print(gpt4 == bard)
```
