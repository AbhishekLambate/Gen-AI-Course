## Abstraction

### Abstraction is the concept of hiding the implmentation details and only showing the necessory features of the object
### This helps in reducing programming complexity and efforts

### Abstraction allows us to work with an object with what an object does rather that how it does it

## How to acheive abstraction in pyton

## 1. Abstract class -> A class that cannot be instantiated or that means for an abstract class we cant create the object of this class.

## 2. Abstract method ->  Abstract method just because it cant accessed by the object of the abstract class itself thats why we have to call this abstract method in our child class and the implementation of the abstract method is not provided inside the abstract class but must be provided in the child class. 

## The abstract method is being defined by @abstractmethod

```python
from abc import ABC, abstractmethod   ## used to create Abstract Base Classes (ABCs)

class Vehicle(ABC):
    
    @abstractmethod
    def start(self):
        pass
    
    @abstractmethod
    def stop(self):
        pass
    
class Car(Vehicle):
    
    def start(self):
        print("The car start with an igition key")
        
    def stop(self):
        print("The car stop with break peddle")
        
class ElectricScooter(Vehicle):
    
    def start(self):
        print("The electric scooter start with power button")
        
    def stop(self):
        print("The electric scooter stop with electric break")
    
```

```python
car=Car()
```

```python
car.start()
```

```python
scooter=ElectricScooter()
```

```python
scooter.start()
```
