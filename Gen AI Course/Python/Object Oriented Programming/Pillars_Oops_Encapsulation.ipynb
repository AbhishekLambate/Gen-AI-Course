## Encapsulation

## Encapsulation -> Its the way of wrapping data (variables and methods) inside a capsule or a class as a single unit.

## It restrics the direct access to some of the objects component's which is a means of preventing accidental inference and misuse of the data.

## Key concepts of Encapsulation:
## 1. Public Members
## 2. Protected Members
## 3. Private Members
## 4. Getter and Setter methods


## Public Members -> Accessible from anywhere or attributes are accessible outside of the class and these members are accessible by the derived class
## Public member attributes are directly declared

```python
class Ironman:
    def __init__(self, name, power, weapon):
        self.name=name
        self.power=power
        self.weapon=weapon
        
    def into(self):
        print(f"Hello My name is {self.name} and I have {self.weapon}")
        
    
    
```

```python
## dir() -> it takes the input as the object name and gives us this idea that which member is public, protected or private
```

```python
iron=Ironman('Ironman', 'Genius Intellect', 'Repuslor Beams')
```

```python
dir(iron)
```

```python
iron.name
```

```python
iron.into()
```

## Private Members -> A Private member is not accessible outside the class and not even in the derived class and in order to access these privatr members or to bring some modification on these members we use getter and setter method.
## Private members are declared by using double undersoce __ befor the name

```python
class Info:
    def __init__(self, name, phone_number):
        self.name=name
        self.__phone_number=phone_number
        
    
```

```python
info=Info('Abhi', '56997656')
```

```python
info.name
```

```python
info.phone_number
```

```python
dir(info)
```

```python
## getter function -> It will be helping us out with accessing the private data
```

```python
class Captain:
    def __init__(self, name, address):
        self.name=name
        self.__address=address
    
    
    ## getter method
    def accessing_add(self):
        print(f"My address is {self.__address}")
```

```python
captain=Captain("Captain", "Newyork")
```

```python
captain.name
```

```python
captain.__address
```

```python
captain.accessing_add()
```

```python
## Setter Method -> It is being used for modifying our private data
```

```python
class Animal:
    def __init__(self, name, sound):
        self.name=name
        self.__sound=sound
        
    ## getter  
    def sound_info(self):
        print(f"The animal is going to sound as {self.__sound}")
        

    ## setter
    def sound_modify(self, new_sound):
        self.__sound=new_sound
        print(f"New sound for the animal is {self.__sound}")
    
```

```python
animal=Animal("Bruno", "Woof Woof")
```

```python
animal.name
```

```python
animal.__sound
```

```python
## getter method to have the access over this data 
```

```python
animal.sound_info()
```

```python
animal.sound_modify("meaw meaw")
```

```python
animal.sound_info()
```

## Protected Members -> Protected data is the data which is not accessible outside the class but the protected data is going to be accessible by the derived class or child class

## For having the protected member you can mention your protected member as _attribute or by using a single underscore

```python
class GrandParent:
    def __init__(self, name, locker_key):
        self.name=name
        self._locker_key=locker_key
        
    ## getter method
    def locker_info(self):
        print(f"The locker key info is {self._locker_key}")
```

```python
class Parent(GrandParent):
    def __init__(self, name, locker_key):
        super().__init__(name, locker_key)
        
    def child_revels(self):
        print(f"The locker key info is {self._locker_key}")
```

```python
grandpa=GrandParent("Abhishek", "123453")
```

```python
grandpa.name
```

```python
dir(grandpa)
```

```python
grandpa.locker_key
```

```python
grandpa.locker_info()
```

```python
parent=Parent("Abhishek", "123452")
```

```python
parent._locker_key
```

```python
parent.locker_info()
```

```python
class BankAccount:
    def __init__(self, holder, balance):
        self.holder=holder
        self.__balance=balance
        
    def get_balance(self):
        print(f"The total balance is {self.__balance}")
        
    def deposite(self, amount):
        if amount > 0:
            self.__balance+=amount
            
            return f"Deposited amount is {self.get_balance()}"
        else:
            return "Deposited amount must be positive"
        
    def withdrow(self, withdrow):
        if 0<withdrow <= self.__balance: 
            self.__balance-=withdrow
            return f"Withdrowal {withdrow} is {self.get_balance()}"
        else:
            return "Insufficient funds in the account"
    
```

```python
account=BankAccount("Abhi", 100000)
```

```python
account.holder
```

```python
account.__balance
```

```python
account.get_balance()
```

```python
account.withdrow(10000)
```
