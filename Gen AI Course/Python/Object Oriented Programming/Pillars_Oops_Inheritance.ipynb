## Pillars of Object Oriented Programming

### Inheritance

#### Inheritance is a fundamental concept in object oriented programming because of which the child class is able to inherit the properties OR Attributes and functionalities of the parent class 

#### Just like the child inherit the properties of parents and the parrents is going to inherit the properties from the grandparent and definitly in this hierarchy the grandchild can have some different features from the parent and grandparent as well 

### Types of Inheritance
### 1. Single Inheritance
### 2. Multilevel Inheritance
### 3. Multiple Inheritance

####  Single Inheritance -> The Features or the attributes from one parent class will be getting inherited into one child class.

####  Parent Class (Properties -> Attributes and functionalities)
####      |
####      |
####    Child Class


#### Multilevel Inheritance -> The properties or methods and attributes of grandparent class can be inherited by parent class and then the properties of the parent class will be inherited by the child class. IN heirarchy the Inheritance process is gong to work for us. 

#### Grandparent (Properties)
####      |
####      |
####    Parent (Properties)
####      |
####      |
####    Child 


#### Multiple INheritance -> The properties of two parent class or multiple parent class will be getting inherited into child class 

####    Parent_1 (Properties) + Parent_2 (Properties) + Parent_3 (Properties)
####      |
####      |
####    Child 

```python
class Avengers:
    def __init__(self, name, power):
        self.name=name
        self.power=power
        
    def intro(self):
        print(f"My name is {self.name}")

class Ironman(Avengers):
    def __init__(self, name, power, wepon):
        super().__init__(name, power)
        self.wepon=wepon
        
    def wepon_detail(self):
        print(f"I use {self.wepon}")
        
```

```python
iron= Ironman('Iron Man', 'Genius Intellect', 'Repulse beams')
iron.name
```

```python
iron.wepon
```

```python
iron.wepon_detail()
```

```python
iron.intro()
```

```python
class Grandparent:
    def __init__(self, nationality, baldness, anger_issues):
        self.nationality=nationality
        self.baldness=baldness
        self.anger_issues=anger_issues
        
    def intro(self):
        print(f"My nationality is {self.nationality} and about my anger issues if you want to ask then the answer is: {self.anger_issues} ")


class Parent(Grandparent):
    def __init__(self, nationality, baldness, anger_issues, intelligence):
        super().__init__(nationality, baldness, anger_issues)
        self.intelligence=intelligence
        
    def intelligence_details(self):
        print(f"About my intelligence if you guys want to know that whether if I am intelligent or not then the answer is {self.intelligence}")
        
class Child(Parent):
    def __init__(self, nationality, baldness, anger_issues, intelligence, charm):
        super().__init__(nationality, baldness, anger_issues, intelligence)
        self.charm=charm
        
    def charm_details(self):
        print(f"About my charm if you guys want to know that wheather if I am charming or not then my answer is {self.charm}")
        
        
```

```python
parent1=Parent('Indian', 'Yes', 'No', 'Yes')
```

```python
parent1.intro()
```

```python
child1=Child('Indian', 'No', 'Yes', 'Yes', 'Yes')
```

```python
child1.intro()
```

```python
child1.intelligence_details()
```

```python
child1.charm_details()
```

```python
child1.nationality="Hindu"
```

```python
child1.intro()
```

```python
class Ironmans:
    def __init__(self, name, power, suit, wepon, nationality='American'):
        self.name=name
        self.power=power
        self.suit=suit
        self.wepon=wepon
        self.nationality=nationality
        
    def intro(self):
        print(f"My name is {self.name} and powers are {self.power}")

class   Captain:
    def __init__(self, name, power, suit, wepon, nationality='American'):
        self.name=name
        self.power=power
        self.suit=suit
        self.wepon=wepon
        self.nationality=nationality
        
    def wepon_details(self):
        print(f"I carry the weapon as : {self.wepon}") 

        
          
```

```python
class Hulk(Ironmans, Captain):
    def __init__(self, name, power, suit, wepon, anger_issues, nationality='American'):
        super().__init__(name, power, suit, wepon, nationality)
        self.anger_issues=anger_issues
        
    def anger_desc(self):
        print(f"About the anger issues if I am having or not then the answer is {self.anger_issues}")
        
```

```python
hulk1=Hulk('Hulk', 'Amazing power', 'Mark 85 Ironman Suit', 'Repulsor beams and shield', 'yes', 'American')
```

```python
hulk1.intro()
```

```python
hulk1.wepon_details()
```
