## Polymorphisum

## Polymorphism means MANY FORMS in Greek
## In OOP's, It allows objects of different classes as if they are the object of the same class and it enables reusability and makes programs more flexible.

## Types of Polymorphism:
## 1. Method Overriding (Runtime Polymorphism)
## 2. Method Overloading (Compile Time Polymorphism)
## 3. Operator Overloading

### Method Overriding -> A child class provides a specific implementation of a method that has already been defined into the parent class. This allows Dynamic Selection at Runtime.

```python
class Grandparent:
    def speak(self):
        print("Hello My name is Abhishek")
        
        
class Parent(Grandparent):
    def speak(self):
        print("Hello My name is Junior Abhishek")
        
class Child(Parent):
    def speak(self):
        print("Hello My name is Junior Abhishek Junior")
```

```python
def attribute_provider(objects):
    return(objects.speak())
```

```python
grandpa=Grandparent()
```

```python
attribute_provider(grandpa)
```

```python
parent=Parent()
```

```python
attribute_provider(parent)
```

```python
child=Child()
```

```python
attribute_provider(child)
```

```python
class Shape:
    def area(self):
        return "This method returns the area of the shape"
    

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width=width
        self.height=height
        
    def area(self):
        return self.width * self.height
    
class Circle(Shape):
    def __init__(self, radius):
        self.radius=radius
        
    def area(self):
        return 3.14*self.radius**2
```

```python
shap=Shape()
shap.area()
```

```python
rectangle=Rectangle(6,4)
rectangle.area()
```

```python
circle=Circle(2)
circle.area()
```
