## Types of constructors:
### 1. Default constructor
### 2. Parameterized constructor

#### 1. Default Constructors : It doesn't have the parameters inside it and it is only having the self keyword inside it. It assigns a default value to all of the objects.

#### 2. Parameterized Constructor: It consists of parameters inside it so that every object can have a specific attribute associated to it.
#### It allows us to pass arrguments while creating object

### Types of Methods
### 1. INstance Method : Instance methods operate on object Attributes
### The instance method is going to use self keyword to access and modify object attributes
### These methods are called using an object


### 2. Class Method : It is the method of the whole class or the whole blueprint which will remain the same for all of the objects then such kind of method is being reffered as a class method
### In class method, These methods works with class level attributes.
### THey are going to use cls intead of self
### They are called using the class name as well
### for specifying that the current method is a class method we have to use the abbreviation as @classmethod


### 3. Static Method : We can say that static methods do not use self, cls
### They work like normal function but belong to a class
### The are going to be designed by a abbrevation @staticmethod
### static method can also be called by the classname also

```python
class Avengers:
    def __init__(self, name, power):
        self.name=name
        self.power=power
    
    def show_details(self):
        print(f"My name is {self.name} and i am having power {self.power}")
        
```

```python
Iron_man=Avengers('Iron man', 'Genius Intellect')

Iron_man.show_details()
```

```python
Thor=Avengers('Thor', 'God of thunder')

Thor.show_details()
```

```python
class Avengerss:
    
    origin='Avenger'
    def __init__(self, name, power):
        self.name=name
        self.power=power
        
    @classmethod
    def origin_info(cls):
        print(f"My origin is {cls.origin}")
        
    def self_info(self):
        print(f"My name is {self.name} and my power is {self.power}")
        
    
```

```python
Iron_man=Avengerss('Iron man', 'Genius Intellect')
```

```python
Iron_man.self_info()
```

```python
Iron_man.origin_info()
```

```python
Thor=Avengerss('Thor', 'God of thunder')
```

```python
Thor.self_info()
```

```python
Thor.origin_info()
```

```python
class Avengersss:
    
    @staticmethod
    def mission():
        print("All avengers must protect earth")
        
    origin="Avenger"
    
    def __init__(self, name, power):
        self.name=name
        self.power=power
        
    @classmethod
    def origin_info(cls):
        print(f"The origin for the avengers is {cls.origin}")
        
    def intro(self):
        print(f"My name is {self.name} and having power {self.power}")
```

```python
Iron_man=Avengersss('Iron man', 'Genius Intellect')
```

```python
Iron_man.intro()
```

```python
Iron_man.origin_info()
```

```python
Iron_man.mission()
```

```python
Avengersss.mission()
```

```python
Avengersss.origin_info()
```
