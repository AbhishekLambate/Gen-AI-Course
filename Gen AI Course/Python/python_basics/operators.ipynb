## Operators

### Operators helps in applying operations on the given operands

### Types of Operators:
1. Arithmatic Operators
    Addition(+)
    Substraction(-)
    Product or Multipication(*)
    Division(/)
    Modulus(%)
    Floor(//)
    Exponent(a**b)

2. Comparision Operators
    Greater than (>)
    less than (<)
    Greater than or equal to (>=)
    less than or equl to (<=)
    comparision/ Equality check(==)
    not equal to (!=)

3. Logical Operators
    and -> whether if both conditions are true or not
    or -> whether if any of the operation is true or not
    not -> negation or the inverse the result of the given operation
